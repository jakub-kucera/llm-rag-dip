```python
import os
import pandas as pd
```

```python
all_files = list(os.listdir("files"))
len(all_files)
```

```python
files_lengths = []
files_lines_count = []
for file in all_files:
    with open(f"files/{file}") as f:
        content = f.read()
        files_lengths.append(len(content))
        files_lines_count.append(content.count("\n") + 1)
```

```python
df = pd.DataFrame({
    "file": all_files,
    "length": files_lengths,
    "lines_count": files_lines_count,
})
df.describe()
```

```python
# histogram of lengths
df.hist(bins=30)
```

```python
# get the longest file
long_file = df.loc[df["length"].idxmax()]
print(f"Longest file: {long_file['file']}")
print(f"Length: {long_file['length']}")
with open(f"files/{long_file['file']}") as f:
    print(f.read())
```

```python
# get the shortest file
short_file = df.loc[df["length"].idxmin()]
print(f"Shortest file: {short_file['file']}")
print(f"Length: {short_file['length']}")
with open(f"files/{short_file['file']}") as f:
    print(f.read())
```
