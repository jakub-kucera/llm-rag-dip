```python
import os
import json
```

```python
# list all files with following name format nss_entities_part_{i}.jsonl
files = os.listdir("..")
files = [f for f in files if f.startswith("nss_entities_part_") and f.endswith(".jsonl")]
files = sorted(files, key=lambda x: int(x.split("_")[-1].split(".")[0]))
files
```

```python
duplicates_count = 0
processed_docs = {}
for file in files:
    print(file)
    with open(file) as f:
        for line in f:
            j_line = json.loads(line)
            # print(j_line.keys())
            # break
            if j_line["file"] in processed_docs:
                if "error" in processed_docs[j_line["file"]]:
                    continue
                print("Duplicate file found:", j_line["file"])
                duplicates_count += 1
            else:
                # if "error" in j_line:
                #     print("Error in file:", j_line["file"])
                    # continue
                processed_docs[j_line["file"]] = j_line
print(f"Duplicates count: {duplicates_count}")
```

```python
len(processed_docs)
```

```python
for name, file in processed_docs.items():
    if "error" in file:
        print("Error in file:", name)
        # continue
```

```python
files_w_error = []
# check that all files were processed
for file in os.listdir("/files"):
    if file not in processed_docs:
        print("File not found:", file)
    elif "error" in processed_docs[file]:
        print("Error in file:", file)
        files_w_error.append(file)
```

```python
with open("../nss_entities_merged.jsonl", "w") as f:
    for file in processed_docs.values():
        f.write(json.dumps(file) + "\n")
```

```python
nss_chunked_lemmatised = []
nss_chunked_lemmatised_path = "../nss_chunked_lemmatised.jsonl"
with open(nss_chunked_lemmatised_path, "r") as f:
    for line in f:
        j_line = json.loads(line)
        nss_chunked_lemmatised.append(j_line)
```

```python
for line in nss_chunked_lemmatised:
    # print(line.keys())
    if line["file"] in processed_docs:
        # if not "response" in processed_docs[line["file"]]:
        if not processed_docs[line["file"]]["response"]:
            print("No response in file:", line["file"])
            continue
        line.update(**processed_docs[line["file"]]["response"])
        # break
        # print(processed_docs[line["file"]]["response"])
    # print(line.keys())
    # break
```

```python
with open("../nss_chunked_lemmatised_w_entities.jsonl", "w") as f:
    for line in nss_chunked_lemmatised:
        f.write(json.dumps(line) + "\n")
```
